Buat list berisi nama 5 mahasiswa

```python
mahasiswa = ["nero", "maha", "esa", "ryan", "zizu"]
print("List nama mahasiswa:")
print(mahasiswa)
```

# Tambahkan 1 nama baru menggunakan `.append()`.


```python
mahasiswa.append("kay")
print("\nSetelah menambahkan nama baru:")
print(mahasiswa)
```

# Hapus satu elemen menggunakan `.remove()`.

```python
mahasiswa.remove("nero")
print("\nSetelah menghapus nama nero:")
print(mahasiswa)
```
