definisi list

```python
list = []
```

input 10 data menggunakan 'for'

```python
for i in range(10):
    angka = int(input("Masukkan angka: "))
    list.append:(angka)
```

tampilkan nilai maksimum, minimum, dan rata-rata

```python
print("List:", list)
print("Nilai maksimum:", max(list))
print("Nilai minimum:", min(list))
print("Rata-rata:", sum(list)/len(list))
```
